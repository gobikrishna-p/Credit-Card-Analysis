##NOW CHECKING FOR OUTLIERS PRESENT IN THE DATASET

```python
column_names = data.columns
column_names = list(column_names)
for i in range(len(column_names)):
  print(column_names[i])
```

```python
#REMOVING OUTLIERS USING IQR(INTER QUANTILE RANGE) METHOD
def Remove_Outliers(data,column_name):
  Q1 = data[column_name].quantile(0.25)
  Q2 = data[column_name].mean()
  Q3 = data[column_name].quantile(0.75)

  #Calculating IQR(Inter Quantile Range)
  IQR = Q3 - Q1

  #Calculating the upper & lower limits
  Upper_Limit = Q3 + (1.5*IQR)
  Lower_Limit = Q1 - (1.5*IQR)

  Outliers = data.loc[(data[column_name] > Upper_Limit) | (data[column_name]< Lower_Limit)]

  #Trimming the data
  before = len(data)
  print("No.of Records Before Removing Outliers:",before)
  data = data.loc[(data[column_name] <= Upper_Limit) & (data[column_name] >= Lower_Limit)]
  print("No.of Records After Removing Outliers:",len(data))
  print("No.of Outliers are:",before-len(data))
  return data
```
